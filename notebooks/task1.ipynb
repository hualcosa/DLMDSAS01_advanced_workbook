```python
import matplotlib.pyplot as plt

# Data for the Bernoulli distribution
outcomes = ["For", "Against"]
xi_2 = 0.22
probabilities = [
    0.22, 
    1 - xi_2,
] 

# Colors for the pie chart: Blue for "For", Orange for "Against"
colors = ["#1f77b4", "#ff7f0e"] 

# Create the pie chart
plt.figure(figsize=(8, 6))
plt.pie(probabilities, labels=outcomes, autopct="%1.1f%%", startangle=90, colors=colors)

# Add a title
plt.title("Bernoulli Trial Outcome Distribution")

# Ensure the pie chart is circular
plt.axis("equal")

# save the figure
plt.savefig("task1_visualization.png")

# Show the plot
plt.show()

```
